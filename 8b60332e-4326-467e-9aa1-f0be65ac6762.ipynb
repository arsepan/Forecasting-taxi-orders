#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

## Обучение

### Предисловие

В задаче было сказано - "нужно спрогнозировать количество заказов такси на следующий час", данное условие возможно сделать с помощью горизонта прогнозирования в один шаг (т.е. на один час вперёд), однако это может не покрыть условия возникновения дополнительных транспортных средств в пиковую нагрузку (я говорю о ситуации, когда машины просто не успеют подъехать по определённым причинам). Что в свою очередь скажется на бизнесе - основатели потеряют дополнительную возможность к заработку. Тогда я сделаю задание и **ПОПЫТАЮСЬ** дополнительно спрогнозировать количество заказов на N часов вперёд. Это позволит более гибко и тщательно распланировать количество автомобилей по радиусу охвата аэропорта.

```python
# Создадим горизонт прогназирования для одного шага


def horizon_one_step(data, max_lag, min_roll):
    
    df = data.copy()
    
    df['year'] = df.index.year
    df['month'] = df.index.month
    df['day'] = df.index.day
    df['dayofweek'] = df.index.dayofweek

    for lag in range(1, max_lag + 1):
        df['lag_{}'.format(lag)] = df['num_orders'].shift(lag)

    df['rolling_mean'] = df['num_orders'].shift().rolling(min_roll).mean()
    
    return df
    
    
one_step = horizon_one_step(data, 500, 50)
display(one_step.tail(5))
```

```python
# Разделим выборку на train и test (10%), также добавил валидационную, чтобы проверять модели не на train
train, test_valid = train_test_split(one_step, shuffle=False, test_size=0.2)
train = train.dropna()
valid, test = train_test_split(test_valid, shuffle=False, test_size=0.5)

print(train.index.min(), train.index.max())
print(valid.index.min(), valid.index.max())
print(test.index.min(), test.index.max())
```

```python
# Разделим выборку на признаки и таргет

features_train = train.drop('num_orders', axis=1)
target_train = train['num_orders']
features_valid = valid.drop('num_orders', axis=1)
target_valid = valid['num_orders']
features_test = test.drop('num_orders', axis=1)
target_test = test['num_orders']
```

```python
# Обучим пустые модели и посмотрим какая справляется лучше

### Линейная регрессия
model_lr = LinearRegression().fit(features_train, target_train)

preds_lr = model_lr.predict(features_valid)

rmse_lr = np.sqrt(mean_squared_error(target_valid, preds_lr))

print("RMSE валидационной выборке:", rmse_lr)
```

```python
def one_step_plot(preds_of_model, target_valid, model):

    preds = pd.DataFrame(data=preds_of_model, index=target_valid.index)
    plot_data = target_valid['2018-08-01':'2018-08-13']
    preds = preds['2018-08-01':'2018-08-13']

    plt.figure(figsize=(15, 5))
    plt.plot(plot_data, color='g', label='История')
    plt.plot(preds, color='r', label='Предсказание модели')
    plt.legend()
    plt.xticks(rotation=45)
    plt.title(f'Прогнозирование с горизонтом в один шаг на {model}', fontsize=16)
    plt.xlabel('Дата')
    plt.ylabel('Количество заказов')
    plt.show()

    plt.figure(figsize=(15, 5))
    plt.plot(plot_data['2018-08-13':'2018-08-13T13:00:00'], color='g', label='История')
    plt.plot(preds['2018-08-13T13:00:00':'2018-08-13T13:00:00'], color='r', label='Прогноз на час вперёд', marker='o')
    plt.legend()
    plt.xticks(plot_data['2018-08-13':'2018-08-13T13:00:00'].index, rotation=45)
    plt.title(f'Точечный прогноз на {model} за последний день', fontsize=16)
    plt.xlabel('Дата')
    plt.ylabel('Количество заказов')
    plt.show()
```

### Горизонт предсказаний в N шагов

#### Предисловие

Пытался придумать каким образом возможно сделать подобное и пришёл к выводу, что необходим алгоритм, который итеррационно будет возвращать спрогнозированное значение в начало очереди. Сугубо говоря, 7 час прогнозирования - 10 заказов такси, тогда после прогнозирования для следующего часа модель нам выдала значение 17, и мы его поставим 8 днём, те перед 10 с точки зрения матрицы. Таким образом по циклу алгоритм будет просчитывать количество заказов на N часов часть признаков будут реальные данные, а часть - прогнозы. Однако у данной модели есть огромный недостаток - некорректность предсказаний, которая будет ухудшать нашу метрику с каждым разом всё больше и больше.

P.S. Если есть какие-то библиотеки для построения горизонта в N шагов - буду рад почитать!

#### Первый вариант

Первый вариант, в котором значения записываются в массив, а далее данный масив двигается при добавлении предсказаний, при этом последний жлемент удаляется. Т.Е. мы прогнозируем одно наблюдение на обученной train выборке.

```python
n_steps = data.copy()

for lag in range(1, 500 + 1):
        n_steps['lag_{}'.format(lag)] = n_steps['num_orders'].shift(lag)
```

```python
train, test = train_test_split(n_steps, shuffle=False, test_size=0.1)
train = train.dropna()

print(train.index.min(), train.index.max())
print(test.index.min(), test.index.max())
```

```python
# Разделим выборку на признаки и таргет

features_train = train.drop('num_orders', axis=1)
target_train = train['num_orders']
features_test = test.drop('num_orders', axis=1)
target_test = test['num_orders']
```

```python
model_r = Ridge().fit(features_train, target_train)
preds_r = model_r.predict(features_train)

a = np.array([features_train.iloc[-1, :]])[0, :-1]
b = np.array([int(round(preds_r[-1], 0))])

new_row = np.append(b, a)

n = 50

for i in range(1, n+1):
    
    preds_r = model_r.predict(np.array([new_row]))
    
    a = new_row[0:-1]
    b = np.array([round(preds_r[-1], 0)])
    
    new_row = np.append(b, a)
```

```python
n_step_plot_target = target_test.head(n)
n_step_plot_features = pd.DataFrame(new_row[:n], index=n_step_plot_target.index)

rmse_n = np.sqrt(mean_squared_error(n_step_plot_target, n_step_plot_features))
print("RMSE валидационной выборке:", rmse_n)

plt.figure(figsize=(20, 8))
plt.plot(n_step_plot_features, label='Предскаазания')
plt.plot(n_step_plot_target, label='История')
plt.legend();
```

#### Второй вариант

Здесь алгоритм добавляет в матрицу строки, как новые наблюдения с предсказаниями.

```python
n_steps = data.copy()

for lag in range(1, 500 + 1):
        n_steps['lag_{}'.format(lag)] = n_steps['num_orders'].shift(lag)
```

```python
train, test = train_test_split(n_steps, shuffle=False, test_size=0.1)
train = train.dropna()

print(train.index.min(), train.index.max())
print(test.index.min(), test.index.max())
```

```python
features_train = train.drop('num_orders', axis=1)
target_train = train['num_orders']
features_test = test.drop('num_orders', axis=1)
target_test = test['num_orders']
```

```python


features_train = features_train.astype(int)
n = target_test.shape[0]

for i in range(n):

    model_r = Ridge().fit(features_train, target_train)
    preds_r = model_r.predict(features_train)

    a = np.array([features_train.iloc[-1, :]])[0, :-1]
    b = np.array([int(round(preds_r[-1], 0))])

    new_row = np.append(b, a)
    features_train.loc[i] = new_row
    
    b = pd.Series(b)
    target_train = target_train.append(b)
```

```python
n_step_plot_target = target_test.head(n-1)
n_step_plot_features = pd.DataFrame(a[:n-1], index=n_step_plot_target.index)

rmse_n = np.sqrt(mean_squared_error(n_step_plot_target, n_step_plot_features))
print("RMSE валидационной выборке:", rmse_n)

plt.figure(figsize=(20, 8))
plt.plot(n_step_plot_features, label='Предскаазания')
plt.plot(n_step_plot_target, label='История')
plt.legend();
```
